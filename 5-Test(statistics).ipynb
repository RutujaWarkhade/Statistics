3. Write a python program to create a graph to find relationship between 
the Culmen Length and Culmen Depth.

```python
#to get relation between Culmen Length and Culmen Depth we have draw a scatter plot
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid");
sns.FacetGrid(df,hue="Species",height=4)\
.map(plt.scatter,'Culmen Length (mm)','Culmen Depth (mm)')\
.add_legend()
plt.show()
```

```python
#from scatter plot it is observer that:
#1)Adelie Penguin (Pygoscelis adeliae) species has culmen length between 30 to 42 mm and culmen depth 16 to 21 mm
#2)Chinstrap penguin (Pygoscelis antarctica) species has culmen length between 40 to 51 mm and culmen depth 17 to 21 mm
#3)Gentoo penguin (Pygoscelis papua) species has culmen length between 40 to 60 mm and culmen depth 13 to 17 mm
```

4. Write a python program to create a pairplot of the penguin data set 
and check which penguin species seems to be the most separable. 
Do write your observations.

```python
#to get general Statistics of Penguin data we have to plot a pair plot
import seaborn as sns
plt.close()
sns.set_style("whitegrid")
sns.pairplot(df,hue="Species",height=4)
plt.show()
```

```python
#form above pair plot it is observe that :
# 1)'Adelie Penguin (Pygoscelis adeliae)' species has highest Culmen Length (mm) upto 59mm 
#'Gentoo penguin (Pygoscelis papua)' species has mid range i.e 53mm and 'Chinstrap penguin (Pygoscelis antarctica)' has smallest length i.e 43mm

# 2)'Adelie Penguin (Pygoscelis adeliae)' species has highest Culmen depth (mm) upto 21mm 
#'Gentoo penguin (Pygoscelis papua)' species has almost range i.e 20mm
#'Chinstrap penguin (Pygoscelis antarctica)' has i.e 16mm

# 3)'Adelie Penguin (Pygoscelis adeliae)' species has Flipper Length (mm) 230mm
#'Gentoo penguin (Pygoscelis papua)' species has almost range i.e 215mm
#'Chinstrap penguin (Pygoscelis antarctica)' has i.e 192mm

# 4)'Adelie Penguin (Pygoscelis adeliae)' species has Body Mass (g) 6300g
#'Gentoo penguin (Pygoscelis papua)' species has almost range i.e 5200g
#'Chinstrap penguin (Pygoscelis antarctica)' has i.e 4600g

# 5)'Gentoo penguin (Pygoscelis papua)' species has highest range of Delta 15 N (o/oo) i.e 10 N
#'Adelie Penguin (Pygoscelis adeliae)' species has 9.2 N
#'Chinstrap penguin (Pygoscelis antarctica)' has 8.5 N

# 6)'Chinstrap penguin (Pygoscelis antarctica)' has highest Delta 13 C (o/oo) i.e -24 C
#'Adelie Penguin (Pygoscelis adeliae)' species has -24.7 C
#'Gentoo penguin (Pygoscelis papua)' species has almost -25.8 C

###From it is observe that 'Chinstrap penguin (Pygoscelis antarctica)' species is separable among all 3 species.
```

5. Write a Python program to find the correlation between variables of
